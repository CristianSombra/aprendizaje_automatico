```python
import pandas as pd

df = pd.read_csv("Automobile.csv")
pd.set_option("display.max_rows", None)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df["weight"], df["mpg"], alpha=0.6, c="blue")
plt.xlabel("Peso del automovil")
plt.ylabel("consumo (mpg)")
plt.title("Relacin entre peso y consumo de combustible")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["mpg"], bins=15, color="green", edgecolor="black", alpha=0.7)
plt.xlabel("Consumo (mpg)")
plt.ylabel("Frecuencia")
plt.title("Distribucin del consumo de combustible")
plt.grid(axis="y", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
box = df.boxplot(column="mpg", by="cylinders", grid=False, patch_artist=True,
                 boxprops=dict(color="black"),
                 medianprops=dict(color="red"))

colors = ["lightblue", "lightgreen", "lightyellow", "lightpink", "lightgray"]
for patch, color in zip(box.artists, colors):
    patch.set_facecolor(color)

plt.title("Consumo (mpg) segn nmero de cilindros")
plt.suptitle("")
plt.xlabel("Nmero de cilindros")
plt.ylabel("Consumo (mpg)")
plt.show()

```

```python
# Tu c칩digo aqu칤
matriz = np.random.randint(1, 21, size=(5, 5))

sum_filas = matriz.sum(axis=1) #sumo horizontalmente
sum_columnas = matriz.sum(axis=0) #sumo verticalmente

print("Matriz:\n", matriz)
print("Suma de filas: ", sum_filas)
print("Suma de columnas: ", sum_columnas)
```

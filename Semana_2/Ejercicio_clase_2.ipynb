```python
import pandas as pd, json

# Crear ventas.csv
pd.DataFrame({
    "fecha": ["2025-09-01","2025-09-01","2025-09-02"],
    "producto": ["A","B","A"],
    "cantidad": [2,1,3],
    "precio_unitario": [1500,2000,1500],
}).to_csv("ventas.csv", index=False)

# Crear clientes.json
clientes = [
    {"id": 1, "nombre": "Juan Pérez", "correo_electrónico": "juan@example.com"},
    {"id": 2, "nombre": "María Gómez", "correo_electrónico": "maria@example.com"},
    {"id": 3, "nombre": "Ana Ruiz", "correo_electrónico": "ana@example.com"},
]
with open("clientes.json","w",encoding="utf-8") as f:
    json.dump(clientes, f, ensure_ascii=False, indent=2)

# Crear inventario.xlsx
pd.DataFrame({
    "producto": ["A","B","C"],
    "stock_disponible": [50,20,75],
    "stock_mínimo": [10,5,15],
}).to_excel("inventario.xlsx", index=False)

print("Archivos creados: ventas.csv, clientes.json, inventario.xlsx")
```
